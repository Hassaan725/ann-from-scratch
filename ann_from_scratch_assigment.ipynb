```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
y = x

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 400)
y = 1 / (1 + np.exp(-x))


plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 1 / (1 + e^(-x))', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = 1 / (1 + e^(-x)) (Sigmoid Function)')
plt.legend()
plt.grid(True)

plt.show()
```

```python

def f(x):
    if x < 0:
        return 0
    else:
        return 1

x_values = np.linspace(-10, 10, 400)


y_values = [f(x) for x in x_values]

plt.figure(figsize=(8, 8))
plt.plot(x_values, y_values, label='f(x) = 0 for x < 0, f(x) = 1 for x >= 0', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function: f(x) = 0 for x < 0, f(x) = 1 for x >= 0')
plt.legend()
plt.grid(True)

plt.show()


```

```python
x = np.linspace(-10, 10, 400)

y = np.log(1 + np.exp(x))
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = ln(1 + e^(x))', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x)= ln(1 + e^(x))')
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(-10, 10, 400)
y=np.tanh(x)

plt.figure(figsize=(8,8))
plt.plot(x,y,label='f(x)=tan^-1 (x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = tanh(x)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
x=np.linspace(-10,10,400)
plt.figure(figsize=(8,6))
y = (2 / (1 + np.exp(-2 * x))) - 1

plt.figure(figsize=(8,8))
plt.plot(x,y, label='(2 / (1 + np.exp(-2 * x))) - 1')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('f(x)=tanh(x)=(2/(1+e^-2^x))-1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha= 0.1
def f(x,alpha):
  if x < 0:
    return alpha*(np.exp(x)-1)
  else:
    return x

x_values = np.linspace(-10,10,400)

y_values = [f(x , alpha) for x in x_values]

plt.figure(figsize=(8, 8))
plt.plot(x_values , y_values , label='')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function: f(x)=0 for x<0, f(x)=1 for x>=0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha= 0.1
def f(x,alpha):
  if x < 0:
    return alpha*x
  else:
    return x

x_values = np.linspace(-10,10,400)

y_values = [f(x , alpha) for x in x_values]

plt.figure(figsize=(8, 6))
plt.plot(x_values , y_values , label='')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function: f(x)=0 for x<0, f(x)=1 for x>=0')
plt.legend()
plt.grid(True)
plt.show()
```
